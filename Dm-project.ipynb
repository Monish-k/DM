```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
import re
pd.set_option('display.max_rows', None)
```

```python
original_df = pd.read_csv("/Users/monish/Desktop/startup_funding.csv",skiprows=0)
req_df=original_df.drop(columns=["Remarks","Sr No","SubVertical","InvestmentnType"])
req_df.dropna(inplace=True)
req_df.reset_index(drop = True,inplace = True)
req_df.rename(columns={"City  Location":"City","Date dd/mm/yyyy":"Date","Startup Name":"Startup","Industry Vertical":"Domain","Amount in USD":"Amount"},inplace =True)
req_df
```

```python
for val in req_df["Investors Name"]:
    if "," in val:
        temp = val.split(",")
        replacer = temp[0]
        req_df["Investors Name"].replace(val,replacer,inplace = True)
for val in req_df["Amount"]:
    if "." in val:
        temp = val.split(".")
        replacer = temp[0]
        req_df["Amount"].replace(val,replacer,inplace = True)
for val in req_df["Startup"]:
    if "http" in val:
        temp = val.split(".")
        replacer = temp[1]
        req_df["Startup"].replace(val,replacer,inplace = True)
for val in req_df["City"]:
    if "," in val:
        temp = val.split(",")
        replacer = temp[0]
        req_df["City"].replace(val,replacer,inplace = True)
    if "/" in val:
        temp = val.split("/")
        replacer = temp[0]
        req_df["City"].replace(val,replacer,inplace = True)
    if "and" in val:
        temp = val.split("and")
        replacer = temp[0]
        req_df["City"].replace(val,replacer,inplace = True)
req_df
```
